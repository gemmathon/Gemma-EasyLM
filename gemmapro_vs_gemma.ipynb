```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
gemmapro = AutoModelForCausalLM.from_pretrained("./gemma-ko-2b-dev")
```

```python
gemma = AutoModelForCausalLM.from_pretrained("google/gemma-2b")
```

```python
for name, param in gemma.named_parameters():
    print(name)
```

```python
for name, param in gemmapro.named_parameters():
    print(name)
```

```python
gemmapro_dict = dict(gemmapro.named_parameters())
gemma_dict = dict(gemma.named_parameters())
```

```python
import torch

torch.all(
    gemmapro_dict["model.embed_tokens.weight"]
    == gemma_dict["model.embed_tokens.weight"]
)
```

```python
torch.all(gemmapro_dict["model.norm.weight"] == gemma_dict["model.norm.weight"])
```

```python
# 기존 layer 검사
for i in range(18):
    b = i // 6
    pi = i + b  # pro index
    print(pi, i)
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.self_attn.q_proj.weight"]
            == gemma_dict[f"model.layers.{i}.self_attn.q_proj.weight"]
        )
    )
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.self_attn.k_proj.weight"]
            == gemma_dict[f"model.layers.{i}.self_attn.k_proj.weight"]
        )
    )
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.self_attn.v_proj.weight"]
            == gemma_dict[f"model.layers.{i}.self_attn.v_proj.weight"]
        )
    )
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.self_attn.o_proj.weight"]
            == gemma_dict[f"model.layers.{i}.self_attn.o_proj.weight"]
        )
    )
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.mlp.gate_proj.weight"]
            == gemma_dict[f"model.layers.{i}.mlp.gate_proj.weight"]
        )
    )
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.mlp.up_proj.weight"]
            == gemma_dict[f"model.layers.{i}.mlp.up_proj.weight"]
        )
    )
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.mlp.down_proj.weight"]
            == gemma_dict[f"model.layers.{i}.mlp.down_proj.weight"]
        )
    )
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.input_layernorm.weight"]
            == gemma_dict[f"model.layers.{i}.input_layernorm.weight"]
        )
    )
    print(
        torch.all(
            gemmapro_dict[f"model.layers.{pi}.post_attention_layernorm.weight"]
            == gemma_dict[f"model.layers.{i}.post_attention_layernorm.weight"]
        )
    )
```
